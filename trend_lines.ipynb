## Linhas de tendência

```python
# Importanto biblioitecas

import pandas as pd
import pandas_datareader.data as web
import yfinance as yf
import cufflinks as cf

yf.pdr_override()
```

```python
# Coletando dados - PETR3

ticket = 'PETR3.SA'

start = '2020-1-1'
end = '2022-1-1'

df = web.get_data_yahoo(ticket, start=start, end=end)

df
```

```python
cf.set_config_file(sharing='public',theme='pearl',offline=False)
cf.go_offline()
```

```python
# Adicionando linhas de tendência

qf = cf.QuantFig(df=df.loc["2020"])

qf.add_trendline(date0="2020-1-2", date1="2020-6-30")
```

```python
qf.iplot()
```
