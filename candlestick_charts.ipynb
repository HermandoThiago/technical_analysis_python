```python
import pandas as pd
import pandas_datareader.data as web
import yfinance as yf
import cufflinks as cf

yf.pdr_override()
```

```python
# Coletando dados

symbol = ['TAEE3.SA']

start = '2020-1-1'
end = '2022-1-1'

df = web.get_data_yahoo(symbol, start=start, end=end)
df
```

```python
# GrÃ¡fico de candle

df.loc['2020'].iplot(kind='candle')
```

```python
# GrÃ¡fico OHLC

df.loc['2020'].iplot(kind='ohlc')
```

### Ajustando timeframe dos dados

```python
# Dados semanais

weekly = df.resample("W-fri").agg({
    "Open":"first",
    "High":"max",
    "Low": "min",
    "Close": "last"
})

weekly
```

```python
from pandas.tseries.frequencies import to_offset
```

```python
weekly.index = weekly.index - to_offset("4d")
```

```python
# Plotando grÃ¡fico semanal do mÃªs 5 atÃ© mÃªs 9

weekly.loc["5-2020":"9-2020"].iplot(kind="candle")
```
