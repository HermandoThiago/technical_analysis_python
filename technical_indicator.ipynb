## Indicadores técnicos

```python
import pandas as pd
import pandas_datareader.data as web
import yfinance as yf
import cufflinks as cf

yf.pdr_override()
```

```python
# Coletando dados

symbol = ['TAEE3.SA']

start = '2020-1-1'
end = '2022-1-1'

df = web.get_data_yahoo(symbol, start=start, end=end)
df
```

```python
cf.set_config_file(sharing='public',theme='pearl',offline=False)
cf.go_offline()
```

```python
qf = cf.QuantFig(df=df)
```

```python
qf.iplot(title="TAEE3")
```

```python
# Adicionando média móvel de 20 períodos

qf.add_sma(periods=20)
```

```python
qf.iplot(title="TAEE")
```

```python
# Adicionando médias móveis de 100 dias

qf.add_sma(periods=100)
```

```python
# Adicionado bandas de bollinger

qf.add_bollinger_bands(periods=20, boll_std=2)
```

```python
qf.iplot(title="TAEE3")
```

```python
# Resetando dataframe

qf = cf.QuantFig(df=df.loc["5-2020"])
```

```python
# Adicionando macd e dmi

qf.add_macd()
qf.add_dmi()
```

```python
qf.iplot(title="TAEE - MACD & DMI")
```
